**Pre-Processing of Quaterly Gold Price on World**

```python
df=pd.read_csv('Quarterly_Gold_Price_on_World.csv')
df.head()
```

We don't have date in proper format so we have to bring it in format same as dataframe of daily gold prices to process it in pandas properly.

```python
for index in range(df.shape[0]):
  i=df["Date"][index]
  year=i[3:]
  # print(year)
  ne=""
  # 1979-04-30
  if "Q1" in i:
    ne=year+"-03-31"	
  elif "Q2" in i:
    ne=year+"-06-30"	
  elif "Q3" in i:
    ne=year+"-09-30"
  else:
    ne=year+"-12-31"
  df["Date"][index]=ne
df.head()
```
